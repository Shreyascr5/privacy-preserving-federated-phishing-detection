```python
import pandas as pd 
import numpy as np


import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("default")


df=pd.read_csv("../dataset/processed/phishing_clean.csv")

df.head()
```

```python
#distribution


plt.figure(figsize=(6,4))


df["Result"].value_counts().plot(
    kind="bar"
)
plt.title("Class distribution")


plt.xlabel("Class")
plt.ylabel("Count")
plt.legend()
plt.show()

df["Result"].value_counts(normalize=True)*100
```

```python
plt.figure(figsize=(16,12))

sns.heatmap(
    df.corr(),
    cmap="coolwarm",
    center=0
)

plt.title(
    "Feature Correlation Heatmap"
)

plt.show()
```

```python
corr_target = df.corr()["Result"]

corr_target = corr_target.sort_values(
    ascending=False
)

print(corr_target)
```

```python
corr_target.drop(
    "Result"
).head(10).plot(
    kind="barh"
)

plt.title(
    "Top Features Correlated with Phishing"
)

plt.show()
```

Dataset Overview

The phishing website dataset consists of 11,055 instances and 30 predictive attributes along with one target attribute. All features are numerical and represent characteristics commonly associated with phishing websites, such as URL properties, domain registration details, and web traffic indicators.

Data Quality

The dataset contains no missing values, ensuring that no additional imputation or data cleaning techniques were required prior to model training.

Class Distribution

The dataset exhibits a near-balanced distribution between phishing and legitimate websites, which is beneficial for supervised machine learning and reduces the likelihood of class imbalance affecting model performance.
